# Data cleaning

```python
from numpy import nan
df=df.replace("?",nan)
```

```python
null_values=df.isnull().sum()
null_values=pd.DataFrame(null_values,columns=['null'])
j=1
sum_tot=len(df)
null_values['percent']=null_values['null']/sum_tot
round(null_values*100,3).sort_values('percent',ascending=False)
```

```python
df['native.country'].fillna(df['native.country'].mode()[0], inplace=True)
```

```python
df['workclass'].fillna(df['workclass'].mode()[0], inplace=True)
```

```python
df['occupation'].fillna(df['occupation'].mode()[0], inplace=True)
```

# Exploratory Data Analysis

```python
sns.pairplot(df)
```

```python
fig = plt.figure(figsize=(5,5)) 
sns.boxplot(x="income", y="age", data=df)
plt.show()
```

The box plot gives the age group where the concentration of population is there when the income

is >50k the age group is 35-52 years.

is <=50k the age group is 25-45 years.

```python
fig = plt.figure(figsize=(12,6)) 
axis = sns.countplot(x="workclass", hue="income", data=df).set_title("workclass vs count")
```

The data seems to mainly consist private employees.

In All the workclasses number of people earning less then 50k are more then those earning 50k.

```python
sns.catplot(y="education", hue="income", kind="count",
            palette="pastel", edgecolor=".6",
            data=df);
```

This data mostly consist of people who has education as hs-grad

```python
sns.catplot(y="marital.status", hue="sex", col="income",
                 data=df, kind="count", height=4, aspect=.7);
```

The people with marital status as Married-civ-spouce has highest people with income more then 50k.

```python
sns.countplot(y="occupation", hue="income", data=df);
```

Most of the people who have income more then 50k either have prof-speciality or exec-managerial as occupation

```python
plt.figure(figsize=(10,7))
sns.countplot(x="relationship", hue="income", data=df);
```
